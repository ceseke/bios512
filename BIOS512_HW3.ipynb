### Question 1
Imagine we have a data frame called `data`, with a `type` column. Which one works and why?\
Function 1:
```
group_and_tally <- function(df, column){
    df %>% group_by({{ column }}) %>% tally();
}
group_and_tally(data, type);
```

Function 2:
```
group_and_tally <- function(df, column){
    df %>% group_by(column) %>% tally();
}
group_and_tally(data, type);
```

Fucntion 1 should be the one that works due to the group_by({{ column }}). This is because the brackets use a shorthand notation to make the computer put type and data as column names. With function 2 they use group_by(column) which will make the machine look for a column named column and won't name your columns by the proper name.
